```python
#packages
import pandas as pd
import numpy as np
from datasets import load_dataset
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
pop_df = pd.read_csv("preprocessed data/popquorndata.csv") #loading data
```

```python
socialbiascorpus = pd.read_csv("preprocessed data/socialbiascorpus.csv") #loading data
```

```python
merged_dataset = pd.read_csv("preprocessed data/merged_dataset.csv") #loading data
```

```python
#creating custom palletes for each plot
custom_palette = {
    'woman': '#3b528b',
    'man': '#21918c',
    'nonBinary': '#5ec962',}
hue_order = ['man', 'woman', 'nonBinary']

custom_palette_pop = {
    'Woman': '#3b528b',
    'Man': '#21918c',
    'Non-binary': '#5ec962',
}
#using title font for the plots 
title_font={'fontsize': 14,
 'fontweight' : 'bold',
 'verticalalignment': 'baseline'}
```

```python
#plotting the distribution of gender withing each race group in the social bias inference corpus
sns.set(style="whitegrid")


plt.figure(figsize=(10, 5))
ax = sns.countplot(x='annotatorRace', hue='annotatorGender', data=socialbiascorpus, palette=custom_palette, hue_order=hue_order)

ax.set_yscale('log')

ax.set_title('Log Distribution of Gender within Each Race Group SBIC', fontdict=title_font)
ax.set_xlabel('Race')
ax.set_ylabel('Count')
plt.xticks(fontsize=13)

plt.legend(title='Gender')

plt.show()
```

```python
#plotting the distribution of gender withing each race group in the popquorn data

sns.set(style="whitegrid")


plt.figure(figsize=(10, 5))
ax = sns.countplot(x='race', hue='gender', data=pop_df, palette= custom_palette_pop)

ax.set_yscale('log')

ax.set_title('Log Distribution of Gender within Each Race Group POPQUORN Data', fontdict=title_font)
ax.set_xlabel('Race')
ax.set_ylabel('Count')
plt.xticks(fontsize=13)

plt.legend(title='Gender')


plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style="whitegrid")

plt.figure(figsize=(12, 6))
ax = sns.countplot(x='race', hue='Dataset', data=merged_dataset, palette='viridis')

ax.set_yscale('log')

ax.set_title('Log Distribution of Source Dataset within Each Race Group', fontdict={'fontsize': 16, 'fontweight': 'bold'})
ax.set_xlabel('Race')
ax.set_ylabel('Count')
plt.xticks(fontsize=13)
plt.yticks(fontsize=13)

plt.legend(title='Dataset', title_fontsize='13', fontsize='11')

plt.show()
```

```python
mean_gender_sbic = socialbiascorpus.groupby(['annotatorGender'])['offensiveYN'].mean()
```

```python
mean_gender_pop = pop_df.groupby(['gender'])['offensiveness'].mean()
```

```python
mean_offensiveness_series = pd.Series(mean_gender_sbic)

sns.set(style="whitegrid")

plt.figure(figsize=(12, 6))
ax = sns.barplot(x=mean_gender_sbic.index, y=mean_gender_sbic.values, palette='viridis')
plt.xticks(fontsize=11)

ax.set_title('Average Offensiveness SBIC', fontdict=title_font)
ax.set_xlabel('Gender')
ax.set_ylabel('Average Offensiveness')


plt.show()
```
